#MatPlotLib
This is a begineer's friendly tutorial for learning matplotlib.
Matplotlib is a low level graph plotting library in python. Most of the common utilities of matplotlib lies under pyplot so we will be focusing mostly on that.





```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([100,90,70,40,20,100])
y = np.array([10,70,50,90,80,60])

plt.plot(x,y)            #Draws a line from point to point
                        # Here x axis is the horizontal and y-axis is vertical axis

plt.show()

```

```python

plt.plot(x,y,'H') # We can use various markers like: o, 8, ., ,, x, +, P, H ....
plt.show()

#If we don't specify any axis, it will take 0,1,2.. as default values.

#We can also use fmt - format string with syntax: marker|line|color

plt.plot(y, 'o:r', ms=10, mec='b', mfc='g')         # - Solid Line, : Dotted line, --Dashed, -. Dashed/dotted
                          # We can specify marker size using ms, marker color by mec(markerEdgeColor), mfc(markerFaceColor).
                          #We can use hexadecimal for colors as well.
plt.show()
```

```python
#Lines can be seperately modified using linespace or ls:
plt.plot(x, ls = '-', color='r', lw=3)
plt.plot(y, ls = '--', color='g', lw=3)
plt.show()

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2, ls = '--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.grid(axis='x')
plt.show()
```

#Display Multiple plots.

* subplot(row, column, current_plot): Takes three arguments.

```python
a = [1,2,3,4,5]
b = [1,4,9,16,25]
plt.subplot(1,2,1)
plt.plot(a,b)
plt.title('First Plot')
plt.subplot(1,2,2)
plt.plot(b,a)
plt.title('Second Plot')

plt.suptitle('Multiple Plots')
plt.show()
```

#ScatterPlots
* scatter(): Plots one dot for each observation. Needs two arrays of same length, one for x and another for y values.


```python
import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, c = colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

#Bars
bar(): first and second arguments are the values and then can use color, size,...

barh(): horizontal bar

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50")
plt.show()
```

#Histograms
Used for frequency distributions.

hist(): Only takes one parameter.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```
